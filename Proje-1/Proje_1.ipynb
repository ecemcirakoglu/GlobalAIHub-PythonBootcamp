```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
Names = np.array([])
Last_names = np.array([])
No = np.array([])
Grade = np.array([])
```

```python
count = input("how many student you have ? ")
for i in range(int(count)):
  name = input("enter the name ")
  last_name = input(" enter the last name ")
  no = input("enter the students no ")
  grade = input(" enter the grade ")
  Names = np.append(Names , name)
  Last_names= np.append(Last_names , last_name)
  No = np.append(No , no)
  Grade = np.append( Grade , grade)
```

```python
print(Names)
print(Last_names)
print(No)
print(Grade)
```

```python
dictionary = {"names":Names,
              "last names":Last_names,
              "numbers":No,
              "grades":Grade
    
}
```

```python
Students = pd.DataFrame(dictionary)
Students.head()
#Students.iloc(Students.names)
```

```python
def letter_grade(grade):
    if(100>=grade & grade >= 80):
         return 'A'
    
    elif(80>grade & grade>=65):
        return 'B'
    
    elif(65>grade & grade>=55):
        return 'C'
    
    elif(55>grade & grade>=50):
        return 'D'
    
    else:
        return 'F'

def pass_class(letter_grade):
    if(letter_grade == 'A' or letter_grade == 'B' or letter_grade =='C' or letter_grade =='D'):
       return "Passed"
    else:
       return "Failed"

Students["letter_grades"] = Students["grades"].apply(int).apply(letter_grade) #Harf notları veri setine eklendi
#Students["letter_grades"] 

Students["passed/failed"] = Students["letter_grades"].apply(pass_class) #Harf notlarına bağlı olarak geçme kalma durumuna bakıldı
#Students["passed/failed"]

Students
```

```python
Students['passed/failed'].value_counts().plot.barh()
```

```python
Students['letter_grades'].value_counts().plot.pie()
```
